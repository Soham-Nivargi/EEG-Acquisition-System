```python
y = [[] for i in range(8)]
```

```python
i=0
for line in txt:
    if(line[0] == 's'):
        i=0
        continue
    else:
        y[i].append(int(line.strip()))
        i=i+1
```

```python
def two_complement(c):
   binary_str = bin(c)[2:].zfill(24)
   inverted_num = ''.join('1' if bit == '0' else '0' for bit in binary_str)
   print(inverted_num)
   twos_c = bin(int(inverted_num, 2) + 1)[2:]
   twos_co = twos_c.zfill(24)
   two = int(twos_co, 2)
   two
   return two
```
